```python
import requests
import json

host = "modelmesh-serving.fharo-testing.svc.cluster.local:8008" # Route generated by rhoai.
inference_path = "v2/models/iris-detect-model"
url = f"http://{host}/{inference_path}"
headers = {"Content-Type": "application/json"}

response = requests.get(url, headers=headers)
response_code = response.status_code
data = response.json()
pretty_json = json.dumps(data, indent=2)

print(f'Hitting endpoint ... GET {url}')
print(f'Response Code: {response_code}')
print(pretty_json)
```

```python
data = {
    "inputs": [
        {
            "name": "float_input",
            "shape": [1, 4],
            "datatype": "FP32",
            "data": [[5.1, 3.5, 1.4, 0.2]]
        }
    ]
}

inference_path = "v2/models/iris-detect-model/infer"
url = f"http://{host}/{inference_path}"

print(f'Hitting endpoint ... POST {url}')
response = requests.post(url, headers=headers, json=data)


response_code = response.status_code
data = response.json()
pretty_json = json.dumps(data, indent=2)

print(f'Response Code: {response_code}')
print(pretty_json)
```
